```python
import re
import pandas as pd

output_csv = "filter1.csv"

# Function to extract sentences with keywords and surrounding context
def extract_with_context(text, keywords, word_count=20):
    if pd.isna(text):
        return "Not specified"  # Return default value for NaN values

    # Regular expression to find keywords with surrounding context
    pattern = rf"((?:\S+\s){{0,{word_count}}}?\b(?:{'|'.join(keywords)})\b(?:\s\S+){{0,{word_count}}})"
    
    match = re.search(pattern, text, re.IGNORECASE)
    if match:
        return match.group(1).strip()  # Extract the match and return it
    return "Not specified"  # Default value if no match is found

# Define keywords
keywords = ["year", "experience", "fresher", "Experience Level"]

# Apply the function to the job description column
df['experience_required_filter1'] = df['description'].apply(lambda x: extract_with_context(x, keywords))

# Save the updated DataFrame to a new CSV file
df.to_csv(output_csv, index=False)

print(f"Updated file saved as {output_csv}")
```

```python
import os
import pandas as pd
from groq import Groq
from dotenv import load_dotenv
import time

# Load environment variables
load_dotenv()

# Groq initialization
groq_api_key = os.getenv("GROQ_API_KEY1")
model_name = "llama-3.2-1b-preview"
client = Groq(api_key=groq_api_key)

def extract_experience(client, model_name, description):
    if not description:
        return "No description provided"  # Return a default value for no text

    # Construct the prompt
    prompt = f"""
    Extract the years of experience required from the following job details. If experience is not mentioned explicitly, return 0.

    ### Examples:
    1. "Exp - 5+ Years" → 5
    2. "Must have 7 years of hands-on developer experience." → 7
    3. "6-8 years of relevant experience in software development in skills listed below." → 6 (use the minimum value if a range is given)
    4. "At least five years of experience in a managerial role." → 5
    5. "Experience: None required." → 0
    6. "Proven Experience" → 1
    

    ### Job Text:
    {description}
    """

    messages = [
        {"role": "system", "content": "You are a helpful assistant specializing in extracting years of experience from job details."},
        {"role": "user", "content": prompt},
    ]
    try:
        # Make the request to the Groq API
        response = client.chat.completions.create(
            model=model_name,
            messages=messages,
            temperature=0.5,
            max_tokens=100
        )
        
        # Extract the LLM response
        llm_response = response.choices[0].message.content.strip()
        return llm_response

    except Exception as e:
        print(f"Error processing text:\n{description}\nError: {e}")
        return str(e)  # Return the error message


# Read the CSV file
input_csv = "total_remaining.csv"
# output_csv = "final_result_with_llm_response.csv"

df = pd.read_csv(input_csv)

# Ensure 'description' column exists
if "description" not in df.columns:
    raise ValueError("The input CSV does not have a 'description' column.")

# Add a new column for LLM response
df["llm_response"] = ""

# Process each row
for index, row in df.iterrows():
    description = row["description"]
    id = row.get("id", index)  # Use 'id' if available, otherwise fallback to index
    print(f"Processing ID: {id}")


    start_time = time.time()
    retry_count = 0
    max_retries = 3
    # llm_response = extract_experience(client, model_name, description)
    while retry_count <= max_retries:
        llm_response = extract_experience(client, model_name, description)
        
        if "Error: Rate limit reached" in llm_response:
            retry_count += 1
            print(f"Rate limit reached for ID {id}. Retrying in 3 seconds... (Attempt {retry_count}/{max_retries})")
            time.sleep(3)  # Wait for 3 seconds before retrying
        else:
            break 
    if retry_count > max_retries:
        print(f"Max retries reached for ID {id}. Skipping...")
        llm_response = "Rate limit error after max retries"    
    df.at[index, "llm_response"] = llm_response 
    end_time = time.time() 
    time_taken = end_time - start_time
    print(f"Time taken for ID {id}: {time_taken} seconds")


    # Pause to respect API rate limits
    time.sleep(4)  

# Save the updated DataFrame to a new CSV file
# df.to_csv(output_csv, index=False)

print(f"Processing complete. Updated file saved ")
```
