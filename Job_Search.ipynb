```python
import os
import re
import pandas as pd
from groq import Groq  
from dotenv import load_dotenv
load_dotenv()

# Load the Groq API key from the .env file
groq_api_key = os.getenv('GROQ_API_KEY')

# Initialize the Groq client with the API key
client = Groq(api_key=groq_api_key)

# Load the model name from the environment
model_name = os.getenv("MODEL_NAME")

# Function to interact with the Groq/Llama model
def chatbot_response(client, model_name, messages, temperature=0):
    input_messages = [{"role": msg["role"], "content": msg["content"]} for msg in messages]
    response = client.chat.completions.create(
        model=model_name,
        messages=input_messages,
        temperature=temperature,
        top_p=0.8,
        max_tokens=1000,
    ).choices[0].message.content
    return response

# Function to extract experience from the job description
def extract_experience(description):
    messages = [
        {"role": "system", "content": "You are a helpful assistant that extracts structured data from text."},
        {
            "role": "user",
            "content": (
                "Extract the years of experience required from the following job description. "
                "Only return the number of years of experience (e.g., '5 years') or 'Not specified' "
                "if no experience requirement is mentioned.\n\n" + description
            ),
        },
    ]
    return chatbot_response(client, model_name, messages).strip()

# Function to extract and return sentences containing "year"
def extract_and_print_sentences(text):
    if pd.isna(text):
        return "Not specified"  # Return default value for NaN values
    sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?)\s', text)  # Split into sentences
    for sentence in sentences:
        if re.search(r'\byear\b', sentence, re.IGNORECASE):
            return sentence.strip()
    return "Not specified"  # Return default value if no relevant sentence is found

# Function to process chunks of the CSV file
def process_chunks(input_csv, output_csv, chunk_size=100):
    # Initialize an empty DataFrame to store results
    all_results = []
    
    # Read the CSV in chunks
    for chunk in pd.read_csv(input_csv, chunksize=chunk_size):
        # Apply the function to the description column
        chunk['experience_required'] = chunk['description'].apply(extract_and_print_sentences).apply(extract_experience)
        
        # Append the processed chunk to the result list
        all_results.append(chunk)
    
    # Concatenate all chunks into one DataFrame
    final_df = pd.concat(all_results, ignore_index=True)
    
    # Save the updated DataFrame to a new CSV file
    final_df.to_csv(output_csv, index=False)
    print(f"Updated file saved as {output_csv}")

# File paths
input_csv = "combined_bangalore.csv"
output_csv = "Bangalore_apply_ready.csv"

# Process the file in chunks
process_chunks(input_csv, output_csv, chunk_size=100)
```
