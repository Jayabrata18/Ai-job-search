```python
import os
import pandas as pd
from groq import Groq
from dotenv import load_dotenv
import time

# Load environment variables
load_dotenv()

# Groq initialization
groq_api_key = os.getenv("GROQ_API_KEY1")
model_name = "llama3-8b-8192"
client = Groq(api_key=groq_api_key)

def extract_experience(client, model_name, description):
    if not description:
        return "No description provided"  # Return a default value for no text

    # Construct the prompt
    prompt = f"""
    Extract the years of experience required from the following job details. If experience is not mentioned explicitly, return 0.

    ### Examples:
    1. "Exp - 5+ Years" → 5
    2. "Must have 7 years of hands-on developer experience." → 7
    3. "6-8 years of relevant experience in software development in skills listed below." → 6 (use the minimum value if a range is given)
   
    

    ### Job Text:
    {description}
    """

    messages = [
        {"role": "system", "content": "You are a helpful assistant specializing in extracting years of experience from job details."},
        {"role": "user", "content": prompt},
    ]
    try:
        # Make the request to the Groq API
        response = client.chat.completions.create(
            model=model_name,
            messages=messages,
            temperature=0.5,
            max_tokens=100
        )
        
        # Extract the LLM response
        llm_response = response.choices[0].message.content.strip()
        return llm_response

    except Exception as e:
        print(f"Error processing text:\n{description}\nError: {e}")
        return str(e)  # Return the error message




# Ensure 'description' column exists
if "description" not in df.columns:
    raise ValueError("The input CSV does not have a 'description' column.")

# Add a new column for LLM response
df["llm_response"] = ""

# Process each row
for index, row in df.iterrows():
    description = row["description"]
    id = row.get("id", index)  # Use 'id' if available, otherwise fallback to index
    print(f"Processing ID: {id}")


    start_time = time.time()
    llm_response = extract_experience(client, model_name, description)
    df.at[index, "llm_response"] = llm_response 
    end_time = time.time() 
    time_taken = end_time - start_time
    print(f"Time taken for ID {id}: {time_taken} seconds")


    # Pause to respect API rate limits
    time.sleep(3)  

# Save the updated DataFrame to a new CSV file
df.to_csv("output.csv", index=False)

print(f"Processing complete. Updated file saved as 'output.csv'.")
```

```python
import re
# Define a function to extract the lowest number from a text
def extract_lowest_number(text):
    # Find all numbers in the text
    numbers = [float(num) for num in re.findall(r'\b\d+\.?\d*\b', str(text))]
    # Return the lowest number or None if no numbers are found
    return min(numbers) if numbers else None

# Apply the function to the 'llm_response' column and create a new column 'experience'
df['experience'] = df['llm_response'].apply(extract_lowest_number)
df.head()
```
